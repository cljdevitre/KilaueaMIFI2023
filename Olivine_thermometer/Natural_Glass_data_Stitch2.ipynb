```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import Thermobar as pt
```

## Load in a load of literature data, which is then merged together

```python
Sides_all=pt.import_excel('Compiled_Natural_Glass_Data.xlsx', sheet_name='Sides_MG')
Sides14=Sides_all['Liqs']
Sides14.head()
```

```python
Helz14_all=pt.import_excel('Compiled_Natural_Glass_Data.xlsx', sheet_name='Helz_2014_Glass')
Helz14=Helz14_all['Liqs']
Helz14.head()
```

```python
Helz15_all=pt.import_excel('Compiled_Natural_Glass_Data.xlsx', sheet_name='Helz_2015_Glass')
Helz15f=Helz15_all['Liqs']
Helz15=Helz15f.loc[(Helz15f['FeOt_Liq']>10.4) & (Helz15f['MgO_Liq']>5)]
Helz15.head()
```

```python
Wieser19_all=pt.import_excel('Compiled_Natural_Glass_Data.xlsx', sheet_name='Wieser_2019_Glass')
Wieser19=Wieser19_all['Liqs']
Wieser19.head()
```

```python
Wieser21_all=pt.import_excel('Compiled_Natural_Glass_Data.xlsx', sheet_name='Wieser_2021_Glass')
Wieser21f=Wieser21_all['Liqs']
Wieser21=Wieser21f
Wieser21f.head()
```

```python
Cl_all=pt.import_excel('Compiled_Natural_Glass_Data.xlsx', sheet_name='Clague_in_prep')
Clague=Cl_all['Liqs']

Clague.head()
```

```python
plt.hist(Clague['MgO_Liq'])
```

```python
df_combo=pd.concat([Sides14, Wieser21, Wieser19, Helz14, Helz15, Clague], axis=0)
df_combo_highFe=df_combo.loc[df_combo['MgO_Liq']>5]
```

```python
ms=4
plt.plot(Sides14['MgO_Liq'], Sides14['FeOt_Liq'], 'sk', ms=ms, mfc='grey', label='Sides2014')
plt.plot(Wieser21['MgO_Liq'], Wieser21['FeOt_Liq'], 'ok', ms=ms, label='Wieser2021')
plt.plot(Wieser19['MgO_Liq'], Wieser19['FeOt_Liq'], 'ok', ms=ms, mfc='yellow', label='Wieser2019')
plt.plot(Helz14['MgO_Liq'], Helz14['FeOt_Liq'], '^k',  ms=ms,mfc='orange', label='Helz2014')
plt.plot(Helz15['MgO_Liq'], Helz15['FeOt_Liq'], '^k',  ms=ms,mfc='red', label='Helz2015')
plt.plot(Clague['MgO_Liq'], Clague['FeOt_Liq'], '+g',  ms=ms,mfc='g', label='Clague91,95')
plt.xlabel('MgO (wt%)')
plt.ylabel('FeO (wt%)')
plt.legend()
plt.savefig('Compiled Glasses.png', dpi=200)
#plt.xlim([5, 12])
#plt.ylim([9, 13])
```

## Lets calculate equilibrium ol contents

```python
Temp_Shea=pt.calculate_liq_only_temp(liq_comps=df_combo_highFe, 
                        equationT='T_Shea2022_MgO')
```

## Lets save this dataset for easy training in different workbooks

```python
df_combo_highFe['T_Shea']=Temp_Shea
df_combo_highFe.to_excel('Natural_data_Temp.xlsx')
```
